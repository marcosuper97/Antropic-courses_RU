Запустите следующую ячейку настройки, чтобы загрузить ключ API и установить вспомогательную функцию get_completion.

```python
%pip install anthropic

# Import python's built-in regular expression library
import re
import anthropic

# Retrieve the API_KEY & MODEL_NAME variables from the IPython store
%store -r API_KEY
%store -r MODEL_NAME

client = anthropic.Anthropic(api_key=API_KEY)

# Note that we changed max_tokens to 4K just for this lesson to allow for longer completions in the exercises
def get_completion(prompt: str, system_prompt=""):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=4000,
        temperature=0.0,
        system=system_prompt,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
```

## Урок

**Клод лучше всего реагирует на четкие и прямые инструкции.**

Думайте о Клоде как о любом другом человеке, впервые приступившем к этой работе. **Клод не знает** контекста, что делать, кроме того, что вы ему буквально говорите. Точно так же, как когда вы впервые инструктируете человека по заданию, чем подробнее вы объясните Клоду, чего именно вы хотите, тем лучше и точнее будет ответ Клода».				
				
Если у вас есть сомнения, следуйте **золотому правилу ясных подсказок**:
- Покажите подсказку коллеге или другу и предложите им самим следовать инструкциям, чтобы проверить, смогут ли они дать желаемый результат. Если они в замешательстве, то Клод в замешательстве.

### Примеры

Возьмем такую задачу, как написание стихов. (Игнорируйте любые несоответствия слогов — LLM пока не очень хорошо умеют считать слоги.)

```python
# Prompt
PROMPT = "Write a haiku about robots."

# Print Claude's response
print(get_completion(PROMPT))
```

Это хайку достаточно красивое, но пользователи могут захотеть, чтобы Клод сразу перешел к стихотворению без преамбулы «Вот хайку».

Как нам этого добиться? Мы **просим об этом**!

```python
# Prompt
PROMPT = "Write a haiku about robots. Skip the preamble; go straight into the poem."

# Print Claude's response
print(get_completion(PROMPT))
```

Вот еще один пример. Давайте спросим Клода, кто лучший баскетболист всех времен. Ниже вы можете видеть, что, хотя Клод и перечисляет несколько имен, **он не называет однозначного «лучшего»**.

```python
# Prompt
PROMPT = "Who is the best basketball player of all time?"

# Print Claude's response
print(get_completion(PROMPT))
```

Можем ли мы заставить Клода принять решение и выбрать лучшего игрока? Да! Просто спросите!

```python
# Prompt
PROMPT = "Who is the best basketball player of all time? Yes, there are differing opinions, but if you absolutely had to pick one player, who would it be?"

# Print Claude's response
print(get_completion(PROMPT))
```

### Упражнение 2.1 — испанский язык
Измените `SYSTEM_PROMPT`, чтобы Клод выдавал ответ на испанском языке.

```python
# System prompt - this is the only field you should change
SYSTEM_PROMPT = "[Replace this text]"

# Prompt
PROMPT = "Hello Claude, how are you?"

# Get Claude's response
response = get_completion(PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    return "hola" in text.lower()

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```

### Упражнение 2.3. Напишите историю

Измените `PROMPT` так, чтобы Клод отвечал настолько длинным ответом, насколько вы можете собрать. Если ваш ответ **более 800 слов**, ответ Клода будет оценен как правильный.

```python
# Prompt - this is the only field you should change
PROMPT = "[Replace this text]"

# Get Claude's response
response = get_completion(PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    trimmed = text.strip()
    words = len(trimmed.split())
    return words >= 800

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```
