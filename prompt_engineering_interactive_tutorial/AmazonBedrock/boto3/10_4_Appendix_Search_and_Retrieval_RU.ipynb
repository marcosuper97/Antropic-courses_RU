# Приложение 10.4: Поиск и извлечение

Знаете ли вы, что можете использовать Клода для **поиска себя в Википедии**? Claude может находить и извлекать статьи, после чего вы также можете использовать Claude для их обобщения и синтеза, написания нового контента на основе найденных материалов и многого другого. И не только Википедия! Вы также можете выполнять поиск по своим собственным документам, независимо от того, хранятся ли они в виде обычного текста или встроены в хранилище векторных данных.

Ознакомьтесь с нашими [примерами кулинарной книги RAG] (https://github.com/anthropics/anthropic-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb), чтобы узнать, как дополнить знания Клода и повысить точность и актуальность ответов Клода с помощью данных, полученных из векторных баз данных, Википедии, Интернета и т. д. Там вы также можете узнать, как использовать определенные инструменты [встраивания] (https://docs.anthropic.com/claude/docs/embeddings) и векторные базы данных.

Если вы хотите узнать о передовых архитектурах RAG с использованием Claude, ознакомьтесь с нашими [слайдами технической презентации Claude 3 по архитектурам RAG] (https://docs.google.com/presentation/d/1zxkSI7lLUBrZycA-_znwqu8DDyVhHLkQGScvzaZrUns/edit#slide=id.g2c736259dac_63_782).
