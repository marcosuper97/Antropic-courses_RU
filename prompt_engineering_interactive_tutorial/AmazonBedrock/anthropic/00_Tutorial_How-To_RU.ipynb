### Anthropic SDK и API сообщений
В этом руководстве мы будем использовать [Anthropic python SDK](https://docs.anthropic.com/claude/reference/claude-on-amazon-bedrock) и [Messages API](https://docs.anthropic.com/claude/reference/messages_post).

Ниже приведен пример того, как будет выглядеть запуск приглашения в этом руководстве.

Сначала мы устанавливаем и сохраняем название модели и регион.

```python
import boto3
session = boto3.Session() # create a boto3 session to dynamically get and set the region name
AWS_REGION = session.region_name
print("AWS Region:", AWS_REGION)
MODEL_NAME = "anthropic.claude-3-haiku-20240307-v1:0"

%store MODEL_NAME
%store AWS_REGION
```

Затем мы создаем get_completion — вспомогательную функцию, которая отправляет запрос Клоду и возвращает сгенерированный Клодом ответ. Запустите эту ячейку сейчас.

```python
from anthropic import AnthropicBedrock

client = AnthropicBedrock(aws_region=AWS_REGION)

def get_completion(prompt, system=''):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ],
        system=system
    )
    return message.content[0].text
```

Теперь мы напишем пример приглашения для Клода и распечатаем вывод Клода, запустив нашу вспомогательную функцию get_completion. Запуск ячейки ниже приведет к распечатке ответа от Клода под ней.

Не стесняйтесь экспериментировать со строкой подсказки, чтобы получить разные ответы от Клода.

```python
# Prompt
prompt = "Hello, Claude!"

# Get Claude's response
print(get_completion(prompt))
```

Определенные ранее переменные MODEL_NAME и AWS_REGION будут использоваться на протяжении всего руководства. Просто не забудьте запустить ячейки для каждой страницы руководства сверху вниз.
