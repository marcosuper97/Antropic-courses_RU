```python
%pip install anthropic --quiet

# Import the hints module from the utils package
import os
import sys
module_path = ".."
sys.path.append(os.path.abspath(module_path))
from utils import hints

# Import python's built-in regular expression library
import re
from anthropic import AnthropicBedrock

%store -r MODEL_NAME
%store -r AWS_REGION

client = AnthropicBedrock(aws_region=AWS_REGION)

def get_completion(prompt, system=''):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ],
        system=system
    )
    return message.content[0].text
```

## Урок

Продолжая тему о том, что у Клода нет никакого контекста, кроме того, что вы говорите, иногда важно **побудить Клода сыграть определенную роль (включая весь необходимый контекст)**. Это также известно как подсказка роли. Чем больше деталей в контексте роли, тем лучше.

**Освоение Клодом роли может улучшить его производительность** в самых разных областях: от написания кода и написания кода до подведения итогов. Это похоже на то, как иногда можно помочь людям, когда им говорят «думать как ______». Ролевые подсказки также могут изменить стиль, тон и манеру ответа Клода.

**Примечание.** Запрос роли может происходить либо в системном приглашении, либо в рамках очереди сообщения пользователя.

### Примеры

В приведенном ниже примере мы видим, что без подсказки роли Клод дает **прямой и нестилизованный ответ**, когда его просят рассказать одним предложением о скейтбординге.

Однако, когда мы даем Клоду возможность сыграть роль кота, точка зрения Клода меняется, и, таким образом, **тон ответа, стиль и содержание ответа Клода адаптируются к новой роли**. 

**Примечание.** Дополнительный прием, который вы можете использовать, — это **предоставить контекст Клода целевой аудитории**. Ниже мы могли бы изменить подсказку, чтобы также сообщить Клоду, с кем следует разговаривать. «Ты кот» вызывает совсем другой ответ, чем «ты кот, разговаривающий с толпой скейтбордистов».

Вот приглашение без указания роли в системном приглашении:

```python
# Prompt
PROMPT = "In one sentence, what do you think about skateboarding?"

# Print Claude's response
print(get_completion(PROMPT))
```

Вот тот же вопрос пользователя, но с подсказкой роли.

```python
# System prompt
SYSTEM_PROMPT = "You are a cat."

# Prompt
PROMPT = "In one sentence, what do you think about skateboarding?"

# Print Claude's response
print(get_completion(PROMPT, SYSTEM_PROMPT))
```

Вы можете использовать ролевые подсказки, чтобы заставить Клода подражать определенным стилям письма, говорить определенным голосом или управлять сложностью ответов. **Подсказки по ролям также могут помочь Клоду лучше выполнять математические и логические задачи.**

Например, в приведенном ниже примере есть однозначный правильный ответ: да. Однако Клод ошибается и думает, что ему не хватает информации, но это не так:

```python
# Prompt
PROMPT = "Jack is looking at Anne. Anne is looking at George. Jack is married, George is not, and we don’t know if Anne is married. Is a married person looking at an unmarried person?"

# Print Claude's response
print(get_completion(PROMPT))
```

А что, если мы **заставим Клода действовать как логический бот**? Как это изменит ответ Клода? 

Оказывается, с этим новым ролевым назначением Клод все делает правильно. (Хотя, в частности, не по всем правильным причинам)

```python
# System prompt
SYSTEM_PROMPT = "You are a logic bot designed to answer complex logic problems."

# Prompt
PROMPT = "Jack is looking at Anne. Anne is looking at George. Jack is married, George is not, and we don’t know if Anne is married. Is a married person looking at an unmarried person?"

# Print Claude's response
print(get_completion(PROMPT, SYSTEM_PROMPT))
```

**Примечание.** В ходе этого курса вы узнаете, что существует **множество методов быстрого проектирования, которые можно использовать для получения аналогичных результатов**. Какие методы вы используете, зависит от вас и ваших предпочтений! Мы рекомендуем вам **поэкспериментировать, чтобы найти свой собственный стиль быстрой разработки**.

### Упражнение 3.1 — Математическая коррекция
В некоторых случаях **Клод может испытывать трудности с математикой**, даже с простой математикой. Ниже Клод неверно оценивает математическую задачу как правильно решенную, хотя на втором шаге есть очевидная арифметическая ошибка. Обратите внимание, что Клод на самом деле замечает ошибку, проходя шаг за шагом, но не приходит к выводу, что общее решение неверно.

Измените `PROMPT` и/или `SYSTEM_PROMPT`, чтобы Клод оценил решение как решенное `incorrectly`, а не как правильное.

```python
# System prompt - if you don't want to use a system prompt, you can leave this variable set to an empty string
SYSTEM_PROMPT = ""

# Prompt
PROMPT = """Is this equation solved correctly below?

2x - 3 = 9
2x = 6
x = 3"""

# Get Claude's response
response = get_completion(PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    if "incorrect" in text or "not correct" in text.lower():
        return True
    else:
        return False

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```
